```python
# Imposed Sparsity Baseline — Pruned MLP Construction
## Purpose
Artificially prune trained MLP agents to match KAN sparsity
levels using iterative magnitude pruning with fine-tuning.
Test pruned MLP robustness under the same perturbations used
in 07_robustness_testing.ipynb.

## This Is Experiment 4 — The Mic Drop
If naturally sparse KAN outperforms artificially pruned MLP
at equal sparsity levels, emergent and imposed sparsity are
proven to be distinct generalization mechanisms. This closes
the single biggest counterargument any judge or reviewer
will raise.

## Why Iterative Magnitude Pruning With Fine-Tuning
This is the strongest available MLP pruning method. Weights
with smallest absolute values are removed in rounds, with
fine-tuning after each round to recover performance. Giving
the MLP every possible advantage means nobody can accuse
the comparison of being rigged with a weak baseline.

## Process
1. Take trained MLP agents from notebooks 01 and 02
2. Measure KAN sparsity level for matched condition
3. Prune MLP iteratively until sparsity levels match
4. Fine-tune pruned MLP to recover performance
5. Test pruned MLP under all four perturbation types
6. Compare robustness to naturally sparse KAN at equal sparsity

## Outputs
- Pruned MLP robustness scores across all perturbation types
- Direct comparison table: KAN vs pruned MLP at matched sparsity
- Statistical significance testing with Cohen's d effect sizes
- Visualization of pruning curve showing reward vs sparsity level

## Status: NOT STARTED — Target October 2026
```
